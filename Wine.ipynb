```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import linregress
```

```python
df=pd.read_csv("data/WineQT.csv")
```

```python
df.dtypes
```

```python
df.head()
```

```python
for col in df:
    print(f'{col}: {round(df[col].mean(),2)}')
```

```python
xvalue=df['alcohol']
yvalue=df['quality']


slope,intercept,rvalue,pvalue,stderr=linregress(xvalue,yvalue)
rvalue*rvalue
myline=slope*xvalue+intercept
print(f'r squared = {rvalue*rvalue}')
print(f'p-value = {pvalue}')


plt.scatter(xvalue,yvalue)
plt.plot(xvalue,myline,color='red')
plt.title("Independent variable effect on Dependent")
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variabble')
plt.show()
```
